```python
import numpy as np

# Örnek sinüs sinyali oluşturma
N = 1000   # Sinyal uzunluğu
T = 1.0 / 1000.0   # Örnekleme aralığı
t = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*t)

# Sinyalin Fourier dönüşümünü hesaplama
yf = np.fft.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

# Sonuçları çizdirme
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = plt.imread('/content/drive/MyDrive/RESİM/2023-04-06 (9).png')

# Compute the Fourier transform
img_fft = np.fft.fft2(img)

# Shift the zero frequency component to the center of the spectrum
img_fft_shift = np.fft.fftshift(img_fft)

# Compute the magnitude spectrum
magnitude_spectrum = 20*np.log(np.abs(img_fft_shift))

# Display the magnitude spectrum
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')
plt.show()
```
